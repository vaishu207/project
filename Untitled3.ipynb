```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
def load_images(dataset_path, h=100, w=100):
    X, y, class_names = [], [], []
    person_id = 0
    for person_name in os.listdir(dataset_path):
        person_dir = os.path.join(dataset_path, person_name)
        if os.path.isdir(person_dir):
            class_names.append(person_name)
            for image_name in os.listdir(person_dir):
                img_path = os.path.join(person_dir, image_name)
                img = cv2.imread(img_path)
                if img is None: continue
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                resized = cv2.resize(gray, (w, h)).flatten()
                X.append(resized)
                y.append(person_id)
            person_id += 1
    return np.array(X).T, np.array(y), class_names  # X as mn*p (columns = images)
```

```python
def pca_manual(X, k):
    mean_face = np.mean(X, axis=1, keepdims=True)  # mean vector mn*1
    X_zero_mean = X - mean_face  # mean aligned faces
    C = np.dot(X_zero_mean.T, X_zero_mean)  # surrogate covariance (p x p)
    eigenvalues, eigenvectors = np.linalg.eigh(C)
    idx = np.argsort(-eigenvalues)[:k]
    eigenvectors = eigenvectors[:, idx]
    eigenfaces = np.dot(X_zero_mean, eigenvectors)  # project to original space
    eigenfaces = eigenfaces / np.linalg.norm(eigenfaces, axis=0)  # normalize
    return eigenfaces, mean_face, X_zero_mean
```

```python
def generate_signatures(eigenfaces, X_zero_mean):
    return np.dot(eigenfaces.T, X_zero_mean)
```

```python
def project_face(eigenfaces, mean_face, test_face):
    return np.dot(eigenfaces.T, test_face - mean_face)
```

```python
dataset_path = "C:\\Users\\vaish\\Downloads\\dataset (2)\\dataset\\faces\\"
X, y, class_names = load_images(dataset_path, 100, 100)
X_train, X_test, y_train, y_test = train_test_split(X.T, y, test_size=0.4, random_state=42)
X_train = X_train.T
X_test = X_test.T
```

```python
k = 50  # number of principal components
eigenfaces, mean_face, X_train_zero_mean = pca_manual(X_train, k)
omega_train = generate_signatures(eigenfaces, X_train_zero_mean)
```

```python
ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
ann.fit(omega_train.T, y_train)
```

```python
X_test_zero_mean = X_test - mean_face
omega_test = generate_signatures(eigenfaces, X_test_zero_mean)
y_pred = ann.predict(omega_test.T)
```

```python
acc = accuracy_score(y_test, y_pred)
print(f"Classification Accuracy: {acc * 100:.2f}%")
```

```python
def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(min(len(images), n_row * n_col)):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())

test_titles = [f"Pred: {class_names[pred]}\nTrue: {class_names[true]}" for pred, true in zip(y_pred, y_test)]
plot_gallery(X_test.T, test_titles, 100, 100)
```
